```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_excel('cleaned.xlsx')
df.head()
```

```python
ingredients = df['TranslatedIngredients'].values.tolist()
prepTime = df['PrepTimeInMins'].values.tolist()
cookTime = df['CookTimeInMins'].values.tolist()
totalTime = df['TotalTimeInMins'].values.tolist()
servings = df['Servings'].values.tolist()
cuisine = df['Cuisine'].values.tolist()
course = df['Course'].values.tolist()
diet = df['Diet'].values.tolist()
recipeName = df['RecipeName'].values.tolist()
```

```python
from os import replace
mainIngredientsList = []
toDelete = []
for fullRecipe in ingredients:
    receipe = ''
    receipe = fullRecipe.split(",")
    ingredientsList = ''
    
    
    for item in receipe:                
        if item.count('-')>1:
            item = item.replace('-','')
        item = re.sub(r'[0-9]+',"" ,item). \
                    replace('teaspoons ',"").replace('teaspoon ',"").replace('tsp','').replace('Tsp',''). \
                    replace('tablespoon ',"").replace('tablespoons ',"").replace('tbsp','').replace('Tbsp',''). \
                    replace('cup ','').replace('cups ','').\
                    replace('ml ','').replace('Ml ','').replace('ML ','').replace('mL ',''). \
                    replace('grams ','').replace('gram ','').replace('Grams ','').replace('gms',''). \
                    replace('kg ','').replace('kgs','').replace('kilograms',''). \
                    replace('liter','').\
                    replace('pinch','').replace('inch','').replace('Inch',''). \
                    replace('sprig','').replace('cloves',''). \
                    replace('/','').replace('+','').replace('.','').\
                    replace(' to ','').replace(' or ','').replace('cut','').replace('chopped','').replace('long','').\
                    replace('taste','').replace('Taste','').\
                    replace('Onions','Onion').replace('onions','onion').\
                    replace('Chillies','Chilly').replace('chillies','Chilly').\
                    replace('Tomatoes','Tomato').replace('tomatoes','Tomato')

                    # ratios, hindi
        item = item.split('(',1)[0].split(' - ',1)[0]
        item = item.replace("-","")
        # print(item)
        item = item.strip().title()
        ingredientsList = ingredientsList + item + ', '
    # print(ingredientsList)
    mainIngredientsList.append(ingredientsList[:-2])
# print(mainIngredientsList)
```

```python
print(len(recipeName))
translatedRecipeName = []
for i in range(len(recipeName)):
    isEnglish = True
    count = 0
    item = recipeName[i]
    chars = []
    for c in item:
        if 0 <= ord(c) <= 127:
            continue
        else:
            isEnglish = False   
            count = count + 1   
            chars.append(c)

    if not isEnglish and count > 2:
        # print(item)
        # print(chars)
        translatedRecipeName.append(item.split('- ',1)[-1].split('(',1)[0])

        # print(recipeName[i])
        # print(i)
        # translatedRecipeName.append( item.split('- ',1)[1].split('(',1)[0])
    else:
        translatedRecipeName.append(item)
```

```python
dfnew = pd.DataFrame({
    'recipeName': translatedRecipeName,
    'ingredients': mainIngredientsList,
    'prepTime': prepTime,
    'cookTime': cookTime,
    'totalTime': totalTime,
    'servings': servings,
    'cuisine': cuisine,
    'course': course,
    'diet': diet
})
dfnew.to_excel('new.xlsx')
dfnew.to_csv('indian_food_recipes.csv')
```

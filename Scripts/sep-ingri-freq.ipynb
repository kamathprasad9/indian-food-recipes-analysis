```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv('./indian_food_recipes.csv')
df.head()
```

```python
ingredients = []
text = ''
for i in range(0,len(df)):
    text = df['ingredients'][i].split(', ')
    text = ','.join(text)
    ingredients.append(text)
    text = ' '.join(ingredients)
# print(text)
# pd.DataFrame({
#     'ingredients': [text]
# }).to_csv('ingredients.csv')
```

```python
frequency = {}
for item in text.split(','):
        if (item in frequency):
            frequency[item] += 1
        else:
            frequency[item] = 1
 
ingriList = []
freqList = []
for key, value in frequency.items():
    ingriList.append(key)
    freqList.append(value)

# freqDf = pd.DataFrame.from_dict(frequency)
# freqDf.to_csv('frquency.csv')
# frequency
data = {'ingredient': ingriList,
        'count': freqList} 
pd.DataFrame.from_dict(data).to_csv('frequency.csv')
```

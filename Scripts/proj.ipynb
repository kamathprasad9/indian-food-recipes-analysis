```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_excel('IndianFoodDatasetXLS.xlsx')
df.head()
```

```python
ingredients = df['TranslatedIngredients'].values.tolist()
ingredients[0]
```

```python
mainIngredientsList = []
toDelete = []
for fullRecipe in ingredients:
    receipe = ''
    receipe = fullRecipe.split(",")
    ingredientsList = []
    
    
    for item in receipe:
        isEnglish = True
        for c in item:
            if 0 <= ord(c) <= 127:
                continue
            else:
                # print(str(ingredients.index(fullRecipe)) + 'Hindi')

                isEnglish = False
                toDelete.append(ingredients.index(fullRecipe))                
                break

        if not isEnglish:
            # toDelete.append(True)
            break
                
        if item.count('-')>1:
            item = item.replace('-','')
        item = re.sub(r'[0-9]+',"" ,item). \
                    replace('teaspoons ',"").replace('teaspoon ',""). \
                    replace('tablespoon ',"").replace('tablespoons ',""). \
                    replace('cup ','').replace('cups ','').\
                    replace('ml ','').replace('Ml ','').replace('ML ','').replace('mL ',''). \
                    replace('grams ','').replace('gram ','').replace('Grams ','').replace('gms',''). \
                    replace('kg ','').replace('kgs','').replace('kilograms',''). \
                    replace('/','')

                    # ratios, hindi
        item = item.split('(',1)[0].split(' - ',1)[0]
        # print(item)
        item = item.strip()
        ingredientsList.append(item)
    # print(ingredientsList)
    mainIngredientsList.append(ingredientsList)
# print(mainIngredientsList)
```

```python
toDelete
isHindi = []
for i in range(len(ingredients)):
    if i in toDelete:
        isHindi.append(True)
    else:
        isHindi.append(False)
# 
# df
```

```python

df.insert(5,"ToDelete", isHindi, True)
df
```
